## **Computational graphs**

A computational graph is a directed graph where the nodes correspond to operations or variables. Variables can feed their value into operations, and operations can feed their output into other operations. This way, every node in the graph defines a function of the variables.

* The major difference between TensorFlow and PyTorch is that TensorFlow’s computational graphs are **static** and PyTorch uses **dynamic** computational graphs.

image.png

 
* **TensorFlow (static)** : Build graph once, then run many times.
* **PyTorch (dynamic)** : Each forward pass defines a new graph.

## **Numpy interoperability**

**Numypy interoperability** : PyTorch tensors can be **converted to Numpy arrays** and vice versa efficiently.

**Tensor on CPU** : The returned array **shares an underlying buffer with the tensor storage**. As a result, the Numpy method can be executed effectively at essentially no cost as long as the data sits in CPU RAM, and modifying the Numpy array leads to a change in the originating tensor. 

**Tensor on GPU** : If the tensor is allocated on the GPU, PyTorch makes a copy of the content of the tensor into a Numpy array allocated on the CPU


````python
import torch

points = torch.ones(2, 2)
print(points) #tensor([[1., 1.],[1., 1.]])
points_np = points.numpy() 
print(points_np) #[[1. 1.][1. 1.]]

#convert torch to numpy 
points = torch.from_numpy(points_np) 
print(points) #tensor([[1., 1.],[1., 1.]])
````

## **Numeric types**

Numeric types : The data type specifies the possible values that the tensor can hold numeric types and the number of bytes per value. 
*   torch.float32 or torch.float—32-bit floating-point 
*   torch.float64 or torch.double—64-bit, double floating-point 
*   torch.float16 or torch.half—16-bit, half floating-point  
*   torch.int8—Signed 8-bit integers 
*   torch.uint8—Unsigned 8-bit integers  
*   torch.int16 or torch.short—Signed 16-bit integers  
*   torch.int32 or torch.int—Signed 32-bit integers 
*   torch.int64 or torch.long—Signed 64-bit integers 








````python
import torch

# syntax 1
double_points = torch.ones(10, 2, dtype=torch.double) 
print(double_points.dtype)  
short_points = torch.tensor([1, 2], dtype=torch.short) 
print(short_points.dtype)#torch.int16

# syntax 2
double_points = torch.zeros(10, 2).double() 
print(double_points.dtype) #torch.float64 
short_points = torch.ones(10, 2).short() 
print(short_points.dtype) #torch.int16

# syntax 3
double_points = torch.zeros(10, 2).to(torch.double) 
print(double_points.dtype) #torch.float64 
short_points = torch.ones(10, 2).to(dtype=torch.short) 
print(short_points.dtype) #torch.int16  
````

## **Indexing Tensor** 

**Indexing Tensor (or Slicing Tensor)** : If you need to obtain a tensor that contains all points but the first? That task is easy when you use range indexing notation, the same kind that applies to standard Python lists

````python
import torch


# Indexing 1-d Tensor 

some_list = list(range(6)) 
print("some_list: ",some_list) # [0,1,2,3,4,5]

print("some_list[:]: ", some_list[:]) #[0,1,2,3,4,5]
print("some_list[1:4]: ", some_list[1:4]) #[1,2,3]
print("some_list[1:]: ", some_list[1:]) #[1,2,3,4,5]
print("some_list[:4]: ", some_list[:4]) #[0,1,2,3]
print("some_list[:-1]: ", some_list[:-1]) #[0,1,2,3,4]
print("some_list[1:4:2]: ", some_list[1:4:2])#[1,3]

# Indexing 2-d Tensor

points = torch.tensor([[1,2],[3,4]]) 
print("points: ", points)
print("points[1:]: ", points[1:]) #tensor([[3, 4]]) 
print("points[0:,1:]: ", points[0:, 1:]) #tensor([[2, 4]]) 
print("points[1:,0]: ", points[1:, 0]) #tensor([3])
````

# **Using CPU/GPU**

**Operations and Processing** : There are many operations such as addition, subtraction, multiplication, dot product, and matrix multiplication. All of these operations can be either performed on the CPU/GPU. PyTorch provides a simple function to copy a tensor on the CPU to the GPU. 

* Comparing the performance between matrix multiplication operations on 
the CPU and GPU.


````python
import time 
import torch

a = torch.rand(10000,10000) 
b = torch.rand(10000,10000)
start_time = time.clock()
a.matmul(b)
print(time.clock()-start_time)# 17.536 seconds

a = a.to("cuda")
b = b.to("cuda")
start_time = time.clock()
a.matmul(b)
print(time.clock()-start_time)# 0.0002 seconds

a = a.to("cpu")
b = b.to("cpu")
start_time = time.clock()
a.matmul(b)
print(time.clock()-start_time)# 17.377 seconds
````

# **Linear regression with PyTorch**

**Implementing linear regression** : We use following steps to implement linear regression using PyTorch instead of Numpy.
* Step1 : Prepare a dataset. 

```python
############## Numpy Version ##############
import numpy as np 

x_data = [3.57, 5.59, 5.82, 8.19, 5.63, 4.89, 3.39, 2.18, 4.84, 6.04, 6.84]
y_data = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0] 
```



````python
############## PyTorch Version ##############
import numpy as np
import torch 

x_data = [3.57, 5.59, 5.82, 8.19, 5.63, 4.89, 3.39, 2.18, 4.84, 6.04, 6.84]
y_data = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
 
x_tensor = torch.tensor(x_data)
y_tensor = torch.tensor(y_data) 
````

* Step2 : Implement forward/backward propagation, computing loss functions for linear regression model.  
(model: Y = w*x + b, Loss: mean squared error)

```python
############## Numpy Version ##############
def forward_prop(x, w, b):
    return np.mul(w,x) +b

def compute_cost(predic_y, y): 
    return ((predic_y - y)**2).mean() 

def dloss_fn(predic_y, y): 
    return 2 * (predic_y - y) / predic_y.size(0)  

def backward_prop(x, y, predic_y, w, b):
    d_loss = dloss_fn(predic_y, y)
    dw = d_loss * x  
    db = d_loss * 1.0  
    return np.stack([dw.mean(), db.mean()]) 
``` 

````python
############## PyTorch Version ##############
def forward_prop(x, w, b):
    return torch.mul(w,x).add(b)

def compute_cost(predic_y, y): 
    return torch.sub(predic_y,y).pow(2).mean() 

def dloss_fn(predic_y, y): 
    return torch.sub(predic_y,y).div(predic_y.size(0)).mul(2)
  
def backward_prop(x, y, predic_y, w, b):
    d_loss = dloss_fn(predic_y, y)
    dw = torch.mul(d_loss, x)
    db = torch.mul(d_loss, torch.tensor(1.0))
    return torch.stack([dw.mean(), db.mean()])     
````

* Step3 : build a linear regression model for training
```python
############## Numpy Version ##############
def training_loop(n_epochs, learning_rate, params, x, y):
    for epoch in range(1, n_epochs + 1):
       w, b = params
       predic_y = forward_prop(x, w, b)  
       loss = compute_cost(predic_y, y)
       grad = backward_prop(x, y, predic_y, w, b)  
       params = params - learning_rate * grad  
       if epoch % 100 == 0 :
            print('Epoch %d, Loss %f' % (epoch, float(loss)))     
    return params 
``` 

````python
############## PyTorch Version ##############
def training_loop(n_epochs, learning_rate, params, x, y):
    for epoch in range(1, n_epochs + 1):
       w, b = params 
       predic_y = forward_prop(x, w, b)  
       loss = compute_cost(predic_y, y) 
       grad = backward_prop(x, y, predic_y, w, b) 
       params = torch.sub(params, torch.mul(learning_rate, grad)) 
       if epoch % 100 == 0 :
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
            
    return params 
````

* Step4 :  Run the following code to test your model 

```python

    n_epochs = 1000,

```

*  Step6 (optional) :  After training, you can plot the results by using the following code.

* All code of Numpy version (for comparing) : The Numpy version produced same results with the PyTorch version. 
  

````python
import numpy as np 

x_data = [3.57, 5.59, 5.82, 8.19, 5.63, 4.89, 3.39, 2.18, 4.84, 6.04, 6.84]
y_data = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0] 

def model(x, w, b):
    return torch.mul(w,x) +b

def loss_fn(predic_y, y): 
    return ((predic_y - y)**2).mean() 
  
def dloss_fn(predic_y, y): 
    return 2 * (predic_y - y) / predic_y.size(0) 

def grad_fn(x, y, predic_y, w, b):
    d_loss = dloss_fn(predic_y, y)
    dw = d_loss * x  
    db = d_loss * 1.0  
    return np.stack([dw.mean(), db.mean()])   

def training_loop(n_epochs, learning_rate, params, x, y):
    for epoch in range(1, n_epochs + 1):
        w, b = params

        predic_y = model(x, w, b)  
        loss = loss_fn(predic_y, y)
        grad = grad_fn(x, y, predic_y, w, b)  

        params = params - learning_rate * grad
        if epoch % 100 == 0 :
          print('Epoch %d, Loss %f' % (epoch, float(loss))) 
            
    return params 
 
params = training_loop(
    n_epochs = 1000, 
    learning_rate = 0.1, 
    params = np.asarray([1.0, 0.0]), 
    x = x_tensor, 
    y = y_tensor)  

from matplotlib import pyplot as plt

predic_y = model(x_tensor, *params)  # <1>

fig = plt.figure(dpi=300)
plt.xlabel("Temperature (°Fahrenheit)")
plt.ylabel("Temperature (°Celsius)")
plt.plot(x_tensor.numpy(), predic_y.detach().numpy()) 
plt.plot(x_tensor.numpy(), y_tensor.numpy(), 'o')
plt.savefig("temp_unknown_plot.png", format="png") 
````

# **Autograd and Variables**

**Autograd** : PyTorch tensors can remember where they come from in terms of the operations and parent tensors that originated them, and they can provide the chain of derivatives of such operations with respect to their inputs automatically. You won’t need to derive your model by hand.  

**Tensor** : Tensor and Function are interconnected and build up an acyclic graph, that encodes a complete history of computation. Each tensor has a .grad_fn attribute that references a Function that has created the Tensor. If you want gradients respect to Tensor, you set its attribute requires_grad as True.

**Gradients** : If you want to compute the derivatives, you can call .backward() on a Tensor.  

**Attributes**
* Backward : back-propagation operation on the variable to compute the gradients. When you finish your computation you can call .backward() and have all the gradients computed automatically.  
* grad : The gradient for this tensor will be accumulated into .grad attribute. The gradient of the output with respect to the variables is constructed by automatic differentiation
* grad_fn : Each tensor has a .grad_fn attribute that references a function that has created the Tensor. If the variable x is created by a user, then the function reference is None. In the case of variable y it refers to its function reference. 



````python
from torch.autograd import Variable
import torch 
 
# Create a tensor and set `requires_grad=True`  
x = torch.ones(1, 2, requires_grad=True)
print(x) #tensor([[1., 1.]], requires_grad=True)
 
# Do a tensor operation:
y = x + 2
print(y) #tensor([[3., 3.]], grad_fn=<AddBackward0>)

# `y` was created as a result of an operation 
print(y.grad_fn) #<AddBackward0 object at 0x7f5de7a97ac8>
 
out = y.mean() 
print(out) #tensor(3., grad_fn=<MeanBackward0>)

# Compute gradients   
out.backward()  
# print gradients d(out)/dx 
print(x.grad) #tensor([[0.5000, 0.5000]]) 
````

## **Shallow NN with torch.nn & torch.optim**
* Implementing a shallow NN : We use following steps to implement a shallow neural network using torch.nn and torch.optim. 
 
* Step1 : First, we need to import the PyTorch library.

````python
import torch
import torch.nn as nn
````

* Step2 : Define all the layers and the batch size to start executing the neural network.

````python
# Defining input size, hidden layer size, output size and batch size respectively
n_in, n_h, n_out, batch_size = 10, 5, 1, 10
````

* Step3 : load input data and respective output data.





````python
# Create dummy input and target tensors (data)
x = torch.randn(batch_size, n_in)
y = torch.tensor([[1.0], [0.0], [0.0], [1.0], [1.0], [1.0], [0.0], [0.0], [1.0], [1.0]])
````

* Step4 : Create a sequential model with the help of in-built functions. 

````python
model = nn.Sequential(nn.Linear(n_in, n_h),
 nn.ReLU(),
 nn.Linear(n_h, n_out),
 nn.Sigmoid())
````

* Step5 : Construct the loss function with the help of Gradient Descent optimizer.   

````python
#Construct the loss function
criterion = torch.nn.MSELoss()
# Construct the optimizer (Stochastic Gradient Descent in this case)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
````

* Step6 : Implement the gradient descent model with the iterating loop.



````python
# Gradient Descent
for epoch in range(50):
    
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x)
    
    # Compute and print loss
    loss = criterion(y_pred, y)
    if epoch % 10 == 0 :
        print('epoch: ', epoch,' loss: ', loss.item())
    
    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()

    # perform a backward pass (backpropagation)
    loss.backward()

    # Update the parameters
    optimizer.step()
````
